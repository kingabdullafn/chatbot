```python
import nltk
from nltk.chat.util import Chat, reflections
```

```python

reflections = {

"i am" : "you are",

"i was" : "you were",

"i" : "you",

"i'm" : "you are",

"i'd" : "you would",

"i've" : "you have",

"i'll" : "you will",

"my" : "your",

"you are" : "I am",

"you were" : "I was",

"you've" : "I have",

"you'll" : "I will",

"your" : "my",

"yours" : "mine",

"you" : "me",

"me" : "you"

}
```

```python
pairs = [
    [r"my name is (.*)",
 ["Hello %1, How are you Today ?",]
     ],
  [r"hi|hey|hello",
 ["Hello","Hey there",]
   ],
 [r"what is your name ?",
 ["I am a bot created from Abdullah cooperation you can call me kingabdulla",]
     ],
      [r"Sorry (.*)",
 ["Its alright bruv",]
     ],
     [r"I am fine",
 ["Great to hear that, just tell me when i can help",]
     ],
      [r"(.*) age ?",
 ["Bruh are you serious right now i am a computer program",]
     ],
      [r"what (.*) want",
 ["Make me an offer i bet i can do it",]
     ],
      [r"(.*) created?",
 ["Abdullah created me using Python's NLTK libary","top secret ;)"]
     ],
      [r"(.*)(location|city)",
 ["United kingdom","Birmingham",]
     ],
      [r"how is weather in (.*)",
 ["Weather in %1 is awesome like always","Too hot man her in %1","Too cold in %1 🥶","Brrooo is that a real place"]

     ],
      [r"i work in (.*)",
 ["%1 is an amazing company, didnt hear about but the livng crisis is high nowadays",]
     ],
      [r"(.*)raining in (.*)",
 ["No rain since last week bare dry here in %2","Bare wet here in %2"]
     ],
      [r"how (.*) health(.*)",
 ["I am very healthy thank you for asking",]
     ],
      [r"(.*) (sport|game)",
 ["i am a very big fan of fc mobile",]
     ],
      [r"who (.*) sportsperson?",
 ["Messi","Ronaldo","Roony","Barbar Azam","Salman Ali Agha","Imran khan"]
     ],
      [r"who (.*) (moviestar|actor)",
 ["Chris Hemsworth",]
     ],
      [r"i am looking for online guides and courses for data science, can you suggest a website",
 ["W3 schools is a good website for all ages it cover topics from python to javascript and it is completly free!",]

     ],
      [r"Bye",
 ["Bye see you another time",]
     ],]


def chat():
  print("Hi! I am a chatbot made from Abdullah but you can call me kingabdulla")
  chat=Chat(pairs, reflections)
  chat.converse()

if __name__ == "__main__":
  chat()
```
